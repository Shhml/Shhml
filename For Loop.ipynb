```python
# 1 print elements of a list
a=[1,2,3,4,5]
for i in a:
   print(i,end=" ")
```

```python
#2 sum of elements in a list
a=[1,2,3,4,5]
sum=0
for i in a:
  sum=sum+i
print(sum)
```

```python
#3 factorial calculation
a=int(input("enter a number"))
f=1
for i in range(1,a+1):
  f=f*i
print(f)
```

```python
#6 count vowels in a string
a="hello"
vowels="aeiou"
count=0
for i in a:
  if i in vowels:
    count+=1
print(count)
```

```python
#7 square of numbers
for i in  range(1,11):
  square=i**2
  print(f"square of {i}:{square}")
```

```python
#9 prime number
for num in range(2,51):
  is_prime =True
  for i in range(2,num):
   if num%i==0:
    is_prime= False
    break
  if is_prime:
    print(num)
```
