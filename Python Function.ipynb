# arbitary argument

```python
#arbitary argument
def my_function(*fruits):
  print("my favorite fruit is"+fruits[2])
my_function("orange","mango","apple","banana")
```

```python
#keyword argument
def my_sum(a,b):
  print(a+b)
my_sum(a=5,b=4)
```

```python
#arbitrary keyword argument
def a(**c):
  print(c["a"]+c["b"])
a(a=1,b=2)
```
