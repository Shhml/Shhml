## combine()

```python
#combine()	Compare the values in two DataFrames, and let a function decide which values to keep
import pandas as pd

df1 = pd.DataFrame([[130, 280],
                    [850, 430]])
df2 = pd.DataFrame([[100, 600]
                    [700, 800]])

def myfunc(a, b) :
  if (a.sum() > b.sum()):
    return a
  else:
    return b

print(df1.combine(df2, myfunc))
```

```python
#combine()	Compare the values in two DataFrames, and let a function decide which values to keep
import pandas as pd

df1 = pd.DataFrame([[1,2,3],
                    [13,4,3]])
df2 = pd.DataFrame([[5,6],
                    [7,8]])

def myfunc(a, b) :
  if (a.sum() > b.sum()):
    return a
  else:
    return b

print(df1.combine(df2, myfunc))
```

## 1) using the duplicated() method,find and count the duplicate row in the dataframe based on the'employee id'column

```python
data={"employee ID":[101,102,103,101,104,102],
      "name":["alice","bob","charlie","alice","eve","bob"],
      "department":["hr","finance","it","hr","it","finance"]}
df=pd.DataFrame(data)

newdf=df[df.duplicated(subset="employee ID")]  #here subset used for take a specified column

print("Duplicate rows:")
print(newdf)

df1=newdf.shape[0]     #here shape is used to count there have how many duplicates in zero th index
print(f"Number of duplicate rows:{df1}")
```

```python
import pandas as pd
data={
    'month':['january','february','march'],
    'product a':[10,15,20],
    'product b':[5,6,7]
}
df=pd.DataFrame(data)
newdf=df.drop('month',axis='columns')
print(newdf)
print(newdf.prod())
```

## 11) find duplicates based on the data column

if yoy had a daataframe with a date column,find duplicates based on the date and count how many roes have the same date

```python
data={
     'date':['2024-01-01','2024-01-02','2024-01-01','2024-01-03','2024-01-02'],
    'event':['event1','event2','event3','event4','event5']
}
df=pd.DataFrame(data)

newdf=df[df.duplicated(subset="date")]  #here subset used for take a specified column

print("Duplicate columns:")
print(newdf)

df1=newdf.shape[0]     #here shape is used to count there have how many duplicates in zero th index
print(f"Number of duplicate rows:{df1}")
```

## 12) drop duplicates based on specific column
remove all duplicates rows from the dataframe based on the "employee id" column(keep the first occurrence)

```python
data={"employee ID":[101,102,103,104,101,105,102,106],
      "name":["alice","bob","charlie","david","alice","eve","bob","frank"],
      "department":["hr","finance","it","marketing","hr","finance","finance","hr"],
      "salary":[70000,80000,45000,90000,70000,62000,80000,55000]
      }
df=pd.DataFrame(data)
newdf=df.drop_duplicates(subset="employee ID",keep="first")
print(newdf)
```

## 14)  group by the "category" column and find the sum of "value" for each category

```python
import pandas as pd
df=pd.DataFrame({
    'category':['A','B','A','B','A'],
    'value':[10,20,30,40,50]
})
newdf=df.groupby('category')['value'].sum()
print(newdf)
```

```python
#plotting the 'calories' feature
plt.figure(figsize=(8,5))
plt.plot(df['Calories'],color='red',marker=".")
plt.title('line plot of Calories')
plt.xlabel('index')
plt.ylabel('Calories')
plt.grid(True)
plt.show()
```
