# WHILE LOOP IN PHYTHON


```python
i=0
while i<=10:
  print(i)
  i=i+1
```

# BREAK STATEMENT

```python
i=1
while i<=10:
  print(i)
  if i==5:
    break
  i=i+1
```

# continue statement

```python
i=0
while i<=10:
  i=i+1
  if i==5:
    continue
  print(i)
```
