```python
#BY REFERING KEY
a={"name":"shamil","age":18}
print(a["name"])
```

```python
#GET() METHOD
a={"name":"shamil","age":18}
x=a.get("name")
print(x)
```

```python
#KEYS()
a={"name":"shamil","age":18}
x=a.keys()
print(x)
```

```python
#VALUES()
a={"name":"shamil","age":18}
x=a.values()
print(x)
```

```python
#ITEMS()
a={"name":"shamil","age":18}
x=a.items()
print(x)
```

```python
#ADD / CHANGE ITEMS IN DICT (by reference key)
a={"child1":"shamil","child2":"nahan","child3":"anas"}
a["child4"]="abee"
print(a)
```
